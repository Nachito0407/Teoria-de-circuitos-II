### Función de transferencia

Dada la siguiente función de transferencia

$$G(s) = \frac{4}{s+4}$$

representar matemáticamente la función.

```python
Gtf = control.tf([4],[1,4])
print(Gtf)
w = np.logspace(-1,5,100)
control.bode_plot(Gtf,w,dB=True,Hz=False, wrap_phase=True, grid=True, color="blue")
```
